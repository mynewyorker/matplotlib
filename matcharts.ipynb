```
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y =2*np.cos(x)**5 + 3*np.sin(x)**3

fig,axes=plt.subplots(2,3,figsize=(18,5),facecolor='#ccddef') #定义画布和子图数量
fig.suptitle('Big Title',fontsize=20) #添加整个画布的标题
fig.text(0.2,0.9,'Subtile') #利用text属性添加副标题

axes[0,0].plot(x,y) #折线图
axes[0,1].bar(x,y*y) #柱状图
axes[0,2].hist(y,bins=30) #直方图
axes[1,0].scatter(x,y) #散点图
axes[1,1].barh(x,y) #条形图
axes[1,2].pie([1,2,3,4,5],labels=['A','B','C','D','E']) #饼图

ax1=axes[0,0] #对子图进行详细设置
ax1.set(xlim=[-10,12],ylim=[-6,4],title='This is TU1',xlabel='xlabel',ylabel='ylabel',facecolor='#ffeedd') #设置子图的xy轴范围，子图标题，标签背景颜色等，也可单独使用ax1.set_xlim()进行设置。部分属性不能直接使用set设置
ax1.grid(True) #添加网格
fig.subplots_adjust(left=0.2, bottom=0.1, right=0.8, top=0.8,hspace=0.5) #调整图表位置和间距
plt.show()
```
